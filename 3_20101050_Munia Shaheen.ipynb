```python
import random

def start(no):  #population create
    st=''
    for j in range(4):    #number of gene we want for our problem
        for i in range(int(no)):
            st+=str(random.randint(0,1))
        if j!=3:
            st+=' '
    st=st.split(' ')
    return st            # ['00100111', '00110101', '00010110', '11100110']
                         # returns a randomly chosen set of 4 chromosomes


def crossover(st):   #single point crossover
    new=[]
    l=len(st[0])     #N or the number of batsmen available
    splitter= random.randint(1,6)  #randodm splitter
    for i in range(len(st)):       #number of gene we defined  #0,1,2,3
        for j in range(i+1,len(st)):  #1,2,3  #2,3  #3
            newchild1=st[i][0:splitter]+st[j][splitter:l]    
            newchild2=st[j][0:splitter]+st[i][splitter:l]
            new.append(newchild1)
            new.append(newchild2)
    #print(new)
    return new


def mutation(new,no):
    new2=[]
    for i in new:
        mut=random.randint(0,int(no)-1)  #choosing random index
        if i[mut]=='1':                 
            i=i[:mut]+'0'+i[mut+1:]     #choosing index e value 1 thakle 0 hoye jabe
        elif i[mut]=='0':               
            i=i[:mut]+'1'+i[mut+1:]     #choosing index e value 0 thakle 1 hoye jabe
        new2.append(i)
    return new2



def fitness_check(new2,run,player,total):  #mutation result, list of average run, list of players, target score 
    values=[]
    for i in new2:
        whole = 0
        for j in range(len(i)):
            if i[j]=='1':
                whole+=int(run[j])   #calculating the total score of selected players
        values.append(abs(whole-int(total)))  #fitness function= target score - total score of the selected players
        
    if 0 in values:
        min_val=min(values)    #always will be 0
        minind=values.index(min_val)
        return new2[minind]   #returns the goal gene which is a string datatype
    else:
        max_value=max(values)
        maxind=values.index(max_value)
        new2.pop(maxind)  #removing the offspring who has the worst fitness score
        return new2       #returns a list



    
#--------------------- GA ------------------------

f=open('input()1.txt')
s=f.read()
s=s.split('\n')
player=[]
run=[]
no,total=s[0].split(' ')

for i in range(1,len(s)):
    x,y=s[i].split(' ')
    player.append(x)
    run.append(y)
#print(run, player)
hehe=start(no)  #initial population create
#print(hehe)


limit=1000
for i in range(limit):
    answer=''
    q1= crossover(hehe)
    q2= mutation(q1,no)
    q3 = fitness_check(q2,run,player,total) #receive a string if it has found the perfect goal, else return a list
    if type(q3)==str:
        print(player)
        answer=q3
        print(answer)
        break
    else:
        hehe=q2[0:4] #takes the first 4 genes to crossover again

if answer=='':    #even after trying 1000 times, if we do not get the perfect goal, then return -1
    print(player)
    print('-1')
```
