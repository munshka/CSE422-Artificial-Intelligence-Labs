```python
#id- 20101050

x=open('map.txt', mode = 'r')
y=x.read()

list1=y.splitlines()

list2= []
for i in list1:
    list2+= [i.split()]
#print(list2)

h={} #heuristic

for i in range(len(list2)):
    h[list2[i][0]]= int(list2[i][1])
#print(h) 


neighbour={}
for i in list2:
    #print(i)
    neighbour[i[0]]=[]
    for j in range(2,len(i),2):
        neighbour[i[0]] += [[i[j],int(i[j+1])]]
#print(neighbour)



start= input()
destination = input()

path_cost = {start: 0}    #only the path costs from start to nodes visited
bool= False

def AStarSearch():
    global neighbour, h               
    opened= [[start, h[start]]] 
    closed= [] 

    while True:
        
        fn= [i[1] for i in opened]     
        minimum= min(fn)
        index= fn.index(minimum)          #finding the index of the minimum node in opened
        present_node= opened[index][0]    #Arad, Sibiu, Rimnicu_Vilcea, Fagaras, Pitesti, Bucharest
        closed.append(opened[index])
        #print('opened',opened)

        del opened[index]
        
        if closed[-1][0]== destination:          #break if Goal node is found
            bool= True
            break
    
        for value in neighbour[present_node]:      #the neighbours of the lowest fn valued nodes
            if value[0] in [i[0] for i in closed]: #parent node e jabo na, only child node e jabo
                continue
            
            path_cost.update({value[0]: path_cost[present_node] +  value[1]})  #actual path cost from the start node   
            new_node= path_cost[present_node] + value[1] +  h[value[0]]         #present cost + old cost + heuristic  
            temp= [value[0],new_node]
             
            if temp[0] in opened:              #if value is already visited then choose the smaller one
                val= [i[1] for i in opened]
                c= val.index(temp[1])
                if temp[1]< opened[c][1]:
                    del opened[c]
                    opened.append(temp)
            else:
                opened.append(temp)
    
    
    if bool==False:
        print("NO PATH FOUND")
      
        
        
    else:
        dest = destination                       
        optimal_path = [destination]    
        #print(closed)
        for i in range(len(closed)-2, -1, -1):
            now = closed[i][0]   #Pitesti, Fagaras, Rimnicu_Vilcea, Sibiu, Arad
            #print(now)
       
            if dest in [i[0] for i in neighbour[now]]:
                child_costs = [i[1] for i in neighbour[now]]
                child_nodes = [i[0] for i in neighbour[now]]
                #print(child_costs)
                #print(child_nodes)

                if path_cost[now] + child_costs[child_nodes.index(dest)] == path_cost[dest]:
                    optimal_path.append(now)
                    dest = now
        
                
        optimal_path.reverse()
       

        return closed, optimal_path



closed, optimal_path = AStarSearch()


output= open("output.txt", mode="w")

st=""
st= "Path: "

for i in range(len(optimal_path)-1):
    st+= optimal_path[i] + "-> "

st += optimal_path[-1]
output.write(st)
print(st)

st2=""
st2= "\nTotal Distance: " + str(closed[-1][1])
output.write(st2)
print(st2)   

output.close()
    
    
#Arad
#Bucharest
```
